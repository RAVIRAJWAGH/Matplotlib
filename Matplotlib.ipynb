# Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])
y=np.array([20,10,15,30,50])
```

```python
plt.plot(x,y,color="r",linewidth=2,marker='o',markerfacecolor="orange",markeredgewidth=2,mec="k",markersize=15,linestyle='dashed')
```

```python
plt.plot(x,y,"o",color="blue")  #Plot
```

```python
plt.scatter(x,y,color="blue")   #Scatter
```

```python
import random
```

```python
random.sample(range(20,50),10)
```

```python
#plot 01
plt.subplot(2,2,1)
x1=np.arange(30)
y1=random.sample(range(20,100),30)
plt.plot(x1,y1,"o")

#plot 02
plt.subplot(2,2,3)
x2=np.arange(30)
y2=random.sample(range(20,100),30)
plt.plot(x2,y2,color="r")

#plot 03
plt.subplot(2,2,2)
x3=np.arange(30)
y3=random.sample(range(200,300),30)
plt.plot(x3,y3,linewidth=5,color="green")

#plot 04
plt.subplot(2,2,4)
x4=np.arange(30)
y4=random.sample(range(1000,1500),30)
plt.plot(x4,y4,linestyle="--")


plt.show()
```
