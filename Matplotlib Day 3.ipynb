```python
import numpy as np

# Create sample data
x = np.linspace(0,10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure and two subplots
fig, ax = plt.subplots()

# Plot the data with customizations
ax.plot(x, y1, label='Sin(x)', color='blue', linestyle='--', linewidth=2)
ax.plot(x, y2, label='Cos(x)', color='red', linestyle='-.', linewidth=2)

# Adding labels and a title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Sine and Cosine Functions')

# Adding a legend
ax.legend()

plt.show()
```

```python
x = np.linspace(0,10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.subplot(2,2,1)
plt.plot(x, y1, label='Sin(x)1',color='blue', linestyle='--', linewidth=2)
plt.title("first graph",color="black")

plt.subplot(2,2,2)
plt.plot(x,y2,label='Sin(x)1',color='red', linestyle='--', linewidth=2)
plt.title("Second graph",color="black",)

plt.subplot(2,2,3)
plt.plot(x,y2,label='Sin(x)1',color='yellow', linestyle='--', linewidth=2)
plt.title("third graph",color="black",)

plt.subplot(2,2,4)
plt.plot(x,y2,label='Sin(x)1',color='green', linestyle='--', linewidth=2)
plt.title("fourth graph",color="black",)
```

## Scatter plot

```python
import matplotlib.pyplot as plt

# Create a line plot
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(y=y,x=x)
plt.show()
```

```python
plt.bar(x,y)
plt.show()
```

```python
marks=[10,50,20,30,28,40,37,50,60,70]
intervals=[0,35,70,100]

plt.hist(marks,intervals)
```

```python
sizes = [30, 20, 25, 15]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, data=sizes, autopct='%.2f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some example data
data = [np.random.normal(0, 1, 100),  # Sample 1
        np.random.normal(2, 1, 100),  # Sample 2
        np.random.normal(1, 1, 100)]  # Sample 3

# Create a box plot
plt.boxplot(data, labels=['Sample 1', 'Sample 2', 'Sample 3'])

# Add labels and a title
plt.xlabel('Samples')
plt.ylabel('Values')
plt.title('Box Plot Example')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate example data
data = np.random.rand(5, 5)

# Create a heatmap
plt.imshow(data, cmap='coolwarm', interpolation='nearest')

# Add color bar
plt.colorbar()

# Customize axis labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Heatmap Example')

# Show the plot
plt.show()
```
